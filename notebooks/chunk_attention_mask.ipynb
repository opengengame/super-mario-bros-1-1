```python
import matplotlib.pyplot as plt
import torch

def visualize_causal_mask(T, chunk_size=None):
  """
  Visualizes the causal attention mask.

  Args:
    T: The sequence length.
    chunk_size: The size of chunks for the chunk causal mask.
  """

  mask = torch.tril(torch.ones(T, T))
  plt.imshow(mask, cmap='gray')
  plt.title('Causal Attention Mask')
  plt.xlabel('Target Position')
  plt.ylabel('Source Position')

  if chunk_size is not None:
    # Visualize chunk boundaries
    for i in range(1, T // chunk_size):
      plt.axvline(i * chunk_size - 0.5, color='red', linestyle='--')
      plt.axhline(i * chunk_size - 0.5, color='red', linestyle='--')

    plt.title('Chunk Causal Attention Mask')

  plt.show()

# Example usage:
visualize_causal_mask(T=10)  # Visualize the standard causal mask
visualize_causal_mask(T=12, chunk_size=3)  # Visualize the chunk causal mask
```
